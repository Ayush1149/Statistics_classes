# Standard Deviation in Python & Machine Learning
Detailed exercises on understanding and applying Standard Deviation.

```python
data = [10, 12, 15, 18, 20, 22, 25, 30]
data
```

```python
mean_value = sum(data)/len(data)
mean_value
```

```python
deviations = [(x-mean_value) for x in data]
deviations
```

```python
squared_dev = [(x-mean_value)**2 for x in data]
squared_dev
```

```python
import math
N = len(data)
sample_variance = sum(squared_dev)/(N-1)
sample_sd = math.sqrt(sample_variance)
sample_sd
```

```python
import numpy as np
np.mean(data), np.std(data, ddof=1)
```

## Standardization Example

```python
heights = [150,155,160,162,170,175]
mean_h = np.mean(heights)
std_h = np.std(heights, ddof=1)
[(x-mean_h)/std_h for x in heights]
```
