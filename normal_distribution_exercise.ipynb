### Learning Goals
- Understand the **shape** of normal, left-skewed, and right-skewed distributions  
- Visualize data using `matplotlib` and `seaborn`  
- Apply transformations (log, square) to normalize skewed data


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

## Step 2: Create Dataset - Hours Studied vs Exam Score

```python

data = {
    'Hours_Studied': [2, 3, 3, 4, 4, 5, 5, 5, 6, 6, 6, 7, 7, 7, 8, 8, 9, 9, 10, 11],
    'Exam_Score':    [50, 55, 60, 62, 65, 68, 70, 72, 73, 75, 76, 77, 78, 79, 80, 82, 84, 86, 88, 90]
}
df = pd.DataFrame(data)
df.head()
```

## Step 3: Visualize Normal Distribution

```python

sns.histplot(df['Exam_Score'], kde=True)
plt.title("Normal Distribution Example: Exam Scores")
plt.xlabel("Exam Score")
plt.ylabel("Frequency")
plt.show()
```

## Step 4: Right-Skewed Distribution - Monthly Income

```python

income_data = {
    'Monthly_Income': [15, 18, 20, 22, 25, 28, 30, 32, 35, 38, 40, 50, 60, 75, 90, 120, 150, 200, 250, 300]
}
df_income = pd.DataFrame(income_data)

sns.histplot(df_income['Monthly_Income'], kde=True)
plt.title("Right-Skewed Distribution: Monthly Income")
plt.xlabel("Monthly Income (in thousands)")
plt.ylabel("Frequency")
plt.show()
```

## Step 5: Normalize Right-Skewed Data (Log Transformation)

```python

df_income['Log_Income'] = np.log(df_income['Monthly_Income'])

sns.histplot(df_income['Log_Income'], kde=True)
plt.title("Normalized (Log Transformed) Income Data")
plt.xlabel("Log of Monthly Income")
plt.ylabel("Frequency")
plt.show()
```

## Step 6: Left-Skewed Distribution - Retirement Savings

```python

savings_data = {
    'Savings': [250, 260, 265, 270, 275, 278, 280, 282, 285, 287, 290, 295, 300, 305, 310, 315, 320, 325, 330, 340]
}
df_savings = pd.DataFrame(savings_data)

sns.histplot(df_savings['Savings'], kde=True)
plt.title("Left-Skewed Distribution: Retirement Savings")
plt.xlabel("Savings (in thousands)")
plt.ylabel("Frequency")
plt.show()
```

## Step 7: Normalize Left-Skewed Data (Square Transformation)

```python

df_savings['Squared_Savings'] = df_savings['Savings'] ** 2

sns.histplot(df_savings['Squared_Savings'], kde=True)
plt.title("Normalized (Squared) Savings Data")
plt.xlabel("Squared Savings")
plt.ylabel("Frequency")
plt.show()
```
